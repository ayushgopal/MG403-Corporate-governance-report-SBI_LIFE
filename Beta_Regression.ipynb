```python
mport pandas as p
from pandas import DataFrame as df
import matplotlib.pyplot as pl
from sklearn.linear_model import LinearRegression
import numpy as np

data1=p.read_csv('sbi.csv')
data2=p.read_csv('nifty.csv')
x=df(data2,columns=['Returns'])
y=df(data1,columns=['Returns'])

```

```python
y=y.iloc[::-1]
```

```python
print("nifty", x.values)
```

```python
print("sbi_life", y.values)
```

```python
pl.figure(figsize=(20,10))
pl.scatter(x,y,alpha=0.5)
pl.title('sbi life beta regression')
pl.xlabel('nifty 500 returns')
pl.ylabel('sbi life returns')
regression=LinearRegression()
regression.fit(x,y)
pl.plot(x,regression.predict(x),color='red',linewidth=4)
pl.show()
```

```python
print("a= ", regression.intercept_)
print("beta= ", regression.coef_)
```
